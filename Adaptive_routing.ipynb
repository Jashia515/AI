```python
#creating the edges path from the shortest path i,e shortest path=[2,5,6] path_create=[(2,5),(5,6)]
def path_create(x):
  mp=[]
  for i in range(0,len(x)):
    if len(x)!=i+1:
        mp.append((x[i],x[i+1]))
  return mp
```

```python
#generating requests for 10 journeys using random function
def generate_req(lt,b1,b,G):
  """
  lt is the list of nodes
  b1 signifies hours
  b2 signifies minutes
  G graph
  """
  l1=[]
  for i in range(0,10):
    x=random.sample(lt,2)
    l1.append([nx.shortest_path(G,x[0],x[1],weight='weight',method='dijkstra'),random.sample(b1,1)[0],random.sample(b,1)[0]])
  return l1
```

```python
# this function finds the node with highest increase and the amount of increase
def find_Increse(mp,d4):
  """
  mp = list of edges
  d4 = base load dictionary
  """
  lk=[]
  inc=0
  for j in mp:
      lk.append(d4[j])
  inc_node=mp[lk.index(max(lk))]
  #print(lk)
  inc=sum(lk)/len(lk)
  return inc,inc_node
```

```python

def recur(G,u1,paths,x,d4):
  
  #print("original",x)
  paths=path_create(x)
  #print(paths)
  if len(paths)==1:
    return x
  ic,icn=find_Increse(paths,d4)
  
  visited=icn[0]
  if ic>=0.05:
    x1=nx.shortest_path(G,icn[1],x[-1],weight='weight',method='dijkstra')
    #print("x1=",x1)
    #print("returns",x[:visited]+x1)
    if x1[0]==x[-1]:
      #print("found")
      return x[:visited]+x1
    else:
      #print("------------recursion----------")
      return recur(G,u1,paths,x1,d4)
  else:
    return x
```

```python
# this function takes the path and later finds the shortest path from the increased node to the destination node
def recur_v1(G,u1,paths,x,d4):
  """
  G= graph
  u1= edges, Graph weights, percentage increase in the weight
  """
  #print("original",x)
  paths=path_create(x)
  #print(paths)
  if len(paths)==1:
    return x
  ic,icn=find_Increse(paths,d4)
  
  visited=x.index(icn[0])
  if ic>=0.05:
    x1=nx.shortest_path(G,icn[1],x[-1],weight='weight',method='dijkstra')
    return x[:visited]+x1
  else:
    return x
```

```python
# Changes the weights of the Graph 
def change_weights(G,base_load):
  """
  G=Graph
  base_load= base load dictionary
  """
  u1=[]
  d2={}
  for (u, v) in G.edges():
    p=random.normalvariate(0,0.25)
    n1=G.edges[u,v]['weight']
    #print(n1)
    c=n1*p*base_load
    b=n1*(1+(c/100))
    G.edges[u,v]['weight']=b
    #print(n2)
    u1.append([(u,v),c,b])
    d2[(u,v)]=b
    d2[(v,u)]=b

  return G,u1,d2
```
